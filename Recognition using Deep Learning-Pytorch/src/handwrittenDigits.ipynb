```python
"""
Sequence Number : 2
Ravina Lad
Date : 11th March 2022
CS 5330 Computer Vision
Spring 2021

This Python file includes

- Task 1 : Build and train a network to recognize digits
- Task F : Read the network and run it on the test set
- Task G : Test the network on new inputs

- Task 4 : Design your own experiment
- Task B : Predict the results
- Task C : Execute your plan
"""
import sys
import torch
import torchvision
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt
import cv2
import numpy as np
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, Dataset

#Netwrok Defination
#made changes in Mynetwork and tried to change the parameters
class MyNetwork(nn.Module):
    def __init__(self):
        super(MyNetwork, self).__init__()
        self.conv1 = nn.Conv2d(1, 10, kernel_size=5)
        self.conv2 = nn.Conv2d(10, 20, kernel_size=5)
        self.conv2_drop = nn.Dropout2d()
        self.fc1 = nn.Linear(320, 50)
        self.fc2 = nn.Linear(50, 10)

    def forward(self, x):
        x = F.relu(F.max_pool2d(self.conv1(x), 2))
        x = F.relu(F.max_pool2d(self.conv2_drop(self.conv2(x)), 2))
        x = x.view(-1, 320)
        x = F.relu(self.fc1(x))
        x = F.dropout(x, training=self.training)
        x = self.fc2(x)
        return F.log_softmax(x)

#Modified Network Defination
class UpdatedNetwork(nn.Module):
    def __init__(self):
        super(UpdatedNetwork, self).__init__()
        self.conv1 = nn.Conv2d(1, 15, kernel_size=7)
        self.conv2 = nn.Conv2d(15, 30, kernel_size=7)
        self.conv2_drop = nn.Dropout2d()
        self.fc1 = nn.Linear(750, 50)
        self.fc2 = nn.Linear(50, 10)

    def forward(self, x):
        x = F.relu(F.max_pool2d(self.conv1(x), 2))
        x = F.relu(F.max_pool2d(self.conv2_drop(self.conv2(x)), 2))
        x = x.view(-1, 750)
        x = F.relu(self.fc1(x))
        x = F.dropout(x, training=self.training)
        x = self.fc2(x)
        return F.log_softmax(x)

# main function to run process of loading MNIST data set TASK 1 F and G
def main(argv):

    #declaration of all Hyperparameters used
    batch_size_test = 1000

    test_loader = torch.utils.data.DataLoader(
                  torchvision.datasets.MNIST('mnist', train=False, download=True,
                             transform=torchvision.transforms.Compose([
                                                                      #transforms.Resize([38,38]), uncomment this to run task 4
                                                                      torchvision.transforms.ToTensor(),
                                                                      torchvision.transforms.Normalize(
                                                                      (0.1307,), (0.3081,)) ])),
    batch_size=batch_size_test, shuffle=True)
    examples = enumerate(test_loader)
    batch_idx, (example_data, example_targets) = next(examples) #get next test data

    #repeatable newtwork
    random_seed = 42
    torch.backends.cudnn.enabled = False
    torch.manual_seed(random_seed)

    #Test the network on new inputs (Handwritten Digits)
    continued_network  = MyNetwork()

    #uncomment this to run task 4
    #continued_network = UpdatedNetwork()

    # load the internal state of the network and optimizer when we last saved them.
    network_state_dict = torch.load('/home/ravina/Desktop/OpenCV/MNIST/modelFiles/model.pth')

    #uncomment this to run task 4
    #network_state_dict = torch.load('/home/ravina/Desktop/OpenCV/MNIST/modelFiles/updated_model.pth')
    #print(network_state_dict)
    

    continued_network.load_state_dict(network_state_dict)
    continued_network.eval()
    

    with torch.no_grad():
      output = continued_network(example_data)
    
    #Prediction of digits from examples data Task 1F
    fig = plt.figure()
    for i in range(9):
      plt.subplot(3,3,i+1)
      plt.tight_layout()
      plt.imshow(example_data[i][0], cmap='gray', interpolation='none')
      plt.title("Prediction: {}".format(
        output.data.max(1, keepdim=True)[1][i].item()))
      plt.xticks([])
      plt.yticks([])
    fig

    #task 1F
    #Hand written Data set

    custom_loader = torch.utils.data.DataLoader(
                  torchvision.datasets.ImageFolder('/home/ravina/Desktop/OpenCV/MNIST/data/test/',
                  transform = transforms.Compose([
                              transforms.Resize([28,28]),
                              #transforms.Resize([38,38]), uncomment this to run task 4
                              transforms.Grayscale(num_output_channels=1),
                              transforms.ToTensor(),
                              transforms.RandomInvert(1),
                              torchvision.transforms.Normalize((0.1307,), (0.3081,)),
                              
    ]) ),
    batch_size=batch_size_test, shuffle=True)


    custom_examples = enumerate(custom_loader)
    custom_batch_idx, (custom_example_data, custom_example_targets) = next(custom_examples) #get next test data

    with torch.no_grad():
      custom_output = continued_network(custom_example_data)
    
    #Prediction of digits from examples data Task 1F
    fig = plt.figure()
    for i in range(9):
      plt.subplot(3,3,i+1)
      plt.tight_layout()
      plt.imshow(custom_example_data[i][0], cmap='gray', interpolation='none')
      plt.title("Prediction: {}".format(
        custom_output.data.max(1, keepdim=True)[1][i].item()))
      plt.xticks([])
      plt.yticks([])
    fig
   
    print("Handwritten Data loaded successfully! NEW")
    return

if __name__ == "__main__":
    main(sys.argv)
```
